# PDF to TIFF Converter
 **Kernel:** ArcGIS Pro (arcgispro-py3)
 
 **Purpose:** Converts single-page PDFs to TIFF using `arcpy.conversion.PDFToTIFF`.
 
 Configure the input and output folders in the configuration cell, or set these environment variables before running:
 - `PDF_TO_TIFF_INPUT_DIR`
 - `PDF_TO_TIFF_OUTPUT_DIR`
 - `PDF_TO_TIFF_DPI`


---
 ## Cell 1 - Imports and setup

```python
import arcpy
 import os
 from pathlib import Path
 from datetime import datetime
 
 print(f"ArcPy   : {arcpy.GetInstallInfo()['Version']}")
 print(f"Started : {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
```

---
 ## Cell 2 - Configuration

```python
# User configuration
 # Replace these placeholder paths, or set environment variables before running the notebook.
 
 INPUT_DIR = Path(os.getenv("PDF_TO_TIFF_INPUT_DIR", r"C:\path\to\single_page_pdfs"))
 OUTPUT_DIR = Path(os.getenv("PDF_TO_TIFF_OUTPUT_DIR", str(INPUT_DIR)))
 TIFF_DPI = int(os.getenv("PDF_TO_TIFF_DPI", "300"))
 
 OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
 
 pdf_files = sorted(
     p for p in INPUT_DIR.iterdir()
     if p.is_file() and p.suffix.lower() == ".pdf"
 )
 
 print(f"Input dir  : {INPUT_DIR}")
 print(f"Output dir : {OUTPUT_DIR}")
 print(f"TIFF DPI   : {TIFF_DPI}")
 print(f"PDFs found : {len(pdf_files)}")
 print()
 for p in pdf_files:
     print(f"   {p.name}")
```

---
 ## Cell 3 - Convert PDFs to TIFF

```python
results = {"success": [], "failed": []}
 
 print(f"\n{'='*60}")
 print(f"  Converting {len(pdf_files)} PDF(s) to TIFF @ {TIFF_DPI} DPI...")
 print(f"{'='*60}")
 
 for pdf_path in pdf_files:
     tiff_path = OUTPUT_DIR / f"{pdf_path.stem}.tif"
     try:
         arcpy.conversion.PDFToTIFF(
             in_pdf_file      = str(pdf_path),
             out_tiff_file    = str(tiff_path),
             pdf_password     = "",
             pdf_page_number  = 1,
             pdf_map          = "",
             clip_option      = "NO_CLIP",
             resolution       = TIFF_DPI,
             color_mode       = "RGB_TRUE_COLOR",
             tiff_compression = "LZW",
             geotiff_tags     = "GEOTIFF_TAGS"
         )
         print(f"   OK {tiff_path.name}")
         results["success"].append(tiff_path)
     except Exception as e:
         print(f"   FAILED {pdf_path.name}\n      -> {e}")
         results["failed"].append((pdf_path, str(e)))
 
 print(f"\n{'='*60}")
 print("  SUMMARY")
 print(f"{'='*60}")
 print(f"  Success : {len(results['success'])}")
 print(f"  Failed  : {len(results['failed'])}")
 if results["failed"]:
     print("\n  Failed files:")
     for p, err in results["failed"]:
         print(f"  - {p.name}\n    -> {err}")
 print(f"\n  Output : {OUTPUT_DIR}")
 print(f"  Done   : {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
 print(f"{'='*60}")
```
